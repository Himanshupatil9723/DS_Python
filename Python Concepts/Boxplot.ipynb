```python
#Import neccessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as pe
import warnings
warnings.filterwarnings('ignore')
```

```python
#import cars dataset
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/cars.csv')
df.head()
```

```python
df.describe()
```

```python
#Set figsize of the graph

plt.rcParams['figure.figsize']= (10,5)
plt.rcParams['figure.dpi']= 250
```

# ***Matplotlib Boxplot***

```python
#Boxplot by matplotlib library
plt.boxplot(df['width'])
plt.show()
```

```python
plt.boxplot(df['width'], vert = False , notch = True , patch_artist = True , showmeans = True)
plt.show()
```

```python
df.boxplot()
```

```python
plt.boxplot(df['city-mpg'])
plt.show()
```
