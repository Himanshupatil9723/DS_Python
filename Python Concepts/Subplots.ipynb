# ***Subplot***

```python
#Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## ***Matplotlib Subplot***

```python
x = np.linspace(40,50,50)
y = np.random.random_sample(50)
```

```python
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 250
```

```python
plt.subplot(2,2,1)
plt.scatter(x,y)
plt.title('First')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(2,2,3)
plt.plot(x,y)

plt.subplot(2,2,4)
plt.boxplot(x, vert=False)

plt.subplot(2,2,2)
plt.hist(x,color='yellow')
plt.tight_layout()
```

## ***Matplotlib Subplots***

```python
plt.subplots(2,2)
```

```python
figure,index = plt.subplots(2,2)

index[0,0].scatter(x,y)
index[0,0].set_title('First Graph')
index[0,0].set_xlabel('X-axis')
index[0,0].set_ylabel('Y-axis')

index[0,1].boxplot(x)

index[1,0].plot(x,y)


index[1,1].scatter(x,y)
plt.tight_layout()
```
