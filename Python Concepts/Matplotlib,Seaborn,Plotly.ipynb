***MATPLOTLIB***

```python
#Import library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

Matplotlib Line *plot*

```python
x = np.linspace(10,25,30)
y = np.random.random_sample(30)
```

```python
plt.plot(x,y)
```

```python
#Change the size of the graph
plt.figure(figsize = (10,5),dpi = 200)
plt.plot(x,y)
```

```python
#set graph parameters
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi']= (200)
```

```python
plt.plot(x,y,color = 'red', linestyle = '--')
```

```python
#Multiline graph
a = np.arange(1,20)

log = np.log(a)
sqrt = np.sqrt(a)

plt.plot(a,log,label = 'Log of X', color = 'yellow')
plt.plot(a,sqrt,label = ' sqrt of x', color = 'blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot')
plt.legend()
```

**Matplotlib** **Histogram**

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/cars.csv')
df.head()
```

```python
df.info()
```

```python
#histogram for individual column
plt.hist(df['width'])
```

```python
#mean & median on histogram

median = df['width'].median()
mean = df['width'].mean()
```

```python
plt.hist(df['width'],color = 'red', edgecolor = 'black')
plt.axvline(median,color = 'yellow')
plt.axvline(mean,color = 'green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram for width')
plt.legend()
```

```python
df.hist(color = 'purple',edgecolor = 'black')
plt.tight_layout()
```
